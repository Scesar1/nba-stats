```python
import time
def make_dataframe_avg(player_id, df):
    player = PlayerCareerStats(player_id=player_id, per_mode36="PerGame").get_data_frames()[0]
    player = player.drop(['SEASON_ID', 'LEAGUE_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'PLAYER_AGE'], axis=1)
    player['player_id'] = player_id
    player = player.groupby('PLAYER_ID').mean()
    player = player.round(1)
    df.append(player)
    return player

players = pd.read_csv('players.csv')
temp = players[players['from'] <= 1979]
df = []
for index, row in temp.iterrows():
    print(index)
    player_id = row['player_id']
    make_dataframe_avg(player_id, df)
    
df = pd.concat(df)
print(df)
```

```python
# Rename columns
df.reset_index(inplace=True)
df = df.rename(columns={'GP': 'games', 'GS': 'gs', 'MIN':'mp', 'FGA':'fg', 'FG_PCT':'fg_percent', 'FG3A':'threeP', 'FG3A':'threePA', 'FG3_PCT':'threeP_percent', 'FTA':'ft', 'FT_PCT':'ft_percent', 'OREB':'orb', 'DREB':'drb', 'REB':'trb', 'AST':'ast', 'STL':'stl', 'BLK':'blk', 'TOV':'tov', 'PF':'pf', 'PTS':'pts'})
df
```

```python
players_per_game = pd.read_csv('player_stats_per_game_new.csv')
players = pd.read_csv('players.csv')
players_after_1979 = players[players['from'] > 1979]
id_list = players_after_1979['player_id'].unique()
players_per_game = players_per_game[players_per_game['player_id'].isin(id_list)]
averages_playoffs = []
averages_regular = []
list_of_columns_to_average = ["games","gs","mp","fg","fga","fg_percent","threeP","threePA","three_percent","twoP","twoPA","two_percent","eFG_percent","ft","fta","ft_percent","orb","drb","trb","ast","stl","blk","tov","pf","pts"]
for iden in id_list:
    temp = players_per_game[players_per_game['player_id'] == iden]
    temp = temp.drop(['player_name', 'pos', 'tm', 'year', 'age'], axis=1)
    for col in list_of_columns_to_average:
        temp[col] = temp[col].clip(lower=0)
    grouped = temp.groupby('playoffs_s').mean()
    if 'Y' in grouped.index:
        play_temp = grouped.loc['Y']
        play_temp['playoffs_s'] = 1
        averages_playoffs.append(play_temp.round(1))
    if 'N' in grouped.index:
        reg_temp = grouped.loc['N']
        reg_temp['playoffs_s'] = 0
        averages_regular.append(reg_temp.round(1))
    
playoffs_df = pd.concat(averages_playoffs, axis=1).T.reset_index(drop=True)
regular_season_df = pd.concat(averages_regular, axis=1).T.reset_index(drop=True)
merged_stats = pd.concat([playoffs_df, regular_season_df])
```

```python
merged_stats['player_id'] = merged_stats['player_id'].astype(int)
merged_stats['playoffs_s'] = merged_stats['playoffs_s'].astype(int)
merged_stats_attempt = merged_stats.copy()
merged_stats[merged_stats['player_id'] == 76001]
merged_stats.head(30)
```
