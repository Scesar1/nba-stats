```python
import pandas as pd
awards = pd.read_csv('awards_individual_new.csv')
players = pd.read_csv('players.csv')
players_with_awards = list(awards['player_id'].unique())
players_without_awards = players[~players['player_id'].isin(players_with_awards)]
ids_of_players_without_awards = list(players_without_awards['player_id'].unique())

for i in ids_of_players_without_awards:
    player = players_without_awards[players_without_awards['player_id'] == i]
    temp = pd.DataFrame(columns=awards.columns)
    temp['player_id'] = player['player_id']
    temp['player_name'] = player['player_name']
    awards = pd.concat([awards, temp], ignore_index=True)
 
awards
```

```python
awards.fillna(0, inplace=True)
awards['player_id'] = awards['player_id'].astype(int)
awards['mvp_seasons'] = awards['mvp_seasons'].astype(str)
awards['mvp_count'] = awards['mvp_count'].astype(int)
awards['mvp_seasons'] = awards['mvp_seasons'].astype(str)
awards['dpoy_seasons'] = awards['dpoy_seasons'].astype(str)
awards['smoy_seasons'] = awards['smoy_seasons'].astype(str)
awards['mip_seasons'] = awards['mip_seasons'].astype(str)
awards['mip_count'] = awards['mip_count'].astype(int)
awards['fmvp_seasons'] = awards['fmvp_seasons'].astype(str)
awards['fmvp_count'] = awards['fmvp_count'].astype(int)
awards['smoy_count'] = awards['smoy_count'].astype(int)
awards['dpoy_count'] = awards['dpoy_count'].astype(int)
awards['lg'] = awards['lg'].astype(str)

awards.to_csv('awards_individual_filled.csv', index=False)
```
